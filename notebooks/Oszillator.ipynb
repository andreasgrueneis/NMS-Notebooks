```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Explizites Euler-Verfahren zu Berechnung einer Lösung der Schwingungsgleichung
# d^2 q(t) 
# -------- = -m w q(t)
# d t^2 
#
# mit  q(0) = 1 / m w  und p(0)=0 

#Parameter
m = 1
w = 1

#Paramter für numerische Integration
dt = 0.1 # Schrittweite
tmax = 2*math.pi # Intervallgrenze für Intgration
q0 = 1/m/w # Anfangsbedingung
p0 = 0     # Anfangsbedingung

steps = int(tmax/dt) # Anzahl der notwendigen Integrationsschritte notwendig um Intervallsgrenze zu erreichen

# Initialisierung
q = np.zeros(steps)
q[0] = q0
p = np.zeros(steps)
p[0] = p0
Energie = np.zeros(steps)
Energie[0] = p[0]**2/2./m + m*w**2/ 2. * q[0]**2
# Array der Stützstellen
t = np.arange(steps)*dt

# Numerische Integration
for i in range(1,steps):    
    # Berechnung der Schrittfunktion für q(t)
    Phi_q = p[i-1] / m
    # Schrittweise Integration von Stützstelle zu Stützstelle
    q[i] = q[i-1] + Phi_q * dt
    
    # Berechnung der Schrittfunktion für q(t)
    Phi_p = - m * w**2 * q[i-1]
    # Schrittweise Integration von Stützstelle zu Stützstelle
    p[i] = p[i-1] + Phi_p * dt
    
    Energie[i] = p[i]**2/2./m + m*w**2/ 2. * q[i]**2


# Plot
fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('Explizites Euler-Verfahren')
fig.set_figwidth(8)
fig.set_figheight(16)

ax1.set(xlabel='q (Ort)', ylabel='p (Impuls)')
ax1.plot(q,p,marker='d')
ax2.set(xlabel='t (Zeit)', ylabel='E (Energie)')
ax2.plot(t, Energie)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Implizites Euler-Verfahren zu Berechnung einer Lösung der Schwingungsgleichung
# d^2 q(t) 
# -------- = -m w q(t)
# d t^2 
#
# mit  q(0) = 1 / m w  und p(0)=0 

#Parameter
m = 1.0
w = 1.0

#Paramter
dt = 0.1 # Schrittweite
tmax = 2*math.pi # Intervallgrenze für Intgration
q0 = 1./m/w # Anfangsbedingung
p0 = 0.0     # Anfangsbedingung

steps = int(tmax/dt) # Anzahl der notwendigen Integrationsschritte notwendig um Intervallsgrenze zu erreichen

# Initialisierung
q = np.zeros(steps)
q[0] = q0
p = np.zeros(steps)
p[0] = p0
Energie = np.zeros(steps)
Energie[0] = p[0]**2/2./m + m*w**2/ 2. * q[0]**2
# Array der Stützstellen
t = np.arange(steps)*dt


# Numerische Integration
for i in range(1,steps):    
    # Schrittweise Integration von Stützstelle zu Stützstelle
    q[i] = 1.0/(1.0+dt**2*w**2) * ( q[i-1] + dt * p[i-1]/m )
    # Schrittweise Integration von Stützstelle zu Stützstelle
    p[i] = 1.0/(1.0+dt**2*w**2) * ( p[i-1] - dt * m * w**2. * q[i-1] )
    
    Energie[i] = p[i]**2/2./m + m*w**2/ 2. * q[i]**2


# Plot
fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('Implizites Euler-Verfahren')
fig.set_figwidth(8)
fig.set_figheight(16)

ax1.set(xlabel='q (Ort)', ylabel='p (Impuls)')
ax1.plot(q,p,marker='d')
ax2.set(xlabel='t (Zeit)', ylabel='E (Energie)')
ax2.plot(t, Energie)
```
